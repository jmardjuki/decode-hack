```python
import pandas as pd
import json
import re
import geopandas as gpd
```

```python
def parseJson(jason):
    m = json.loads(jason)['coordinates']
    return m[0], m[1]
```

```python
parkingMeters_df = pd.read_csv("./data/parking-meters.csv", ";", usecols=["PAY_PHONE", "Geom"])
```

```python
parkingMeters_df['lat'], parkingMeters_df['lon'] = zip(*parkingMeters_df['Geom'].map(parseJson))
parkingMeters_df.head()
```

```python
meterLocs = gpd.GeoDataFrame(parkingMeters_df, geometry=gpd.points_from_xy(parkingMeters_df.lat, parkingMeters_df.lon)).set_index("PAY_PHONE")
meterLocs.head()
```

```python
payByPhone_df= pd.read_csv("./data/pay-by-phone-parking-august-1-to-26-2017.csv", ";", usecols=["METER", "DATE START", "MINUTES", "VEHICLE TYPE"])
```

```python
payByPhone_df2= pd.read_csv("./data/no-name1.csv", ";", usecols=["METER", "DATE START", "MINUTES", "VEHICLE TYPE"])
```

```python
payByPhone_df = pd.concat([payByPhone_df, payByPhone_df2])
```

```python
payByPhone_df
```

```python
data = payByPhone_df.join(meterLocs, 'METER').dropna()
data
```

```python
aggdata = data.groupby('METER').agg('count')
aggdata
```
